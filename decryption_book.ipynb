Reading the transmission file

```python
myFile = open("transmission.txt", "r")
dataList = myFile.readlines()
inverted = []
for i in dataList:
    i = i.replace("\n","")
    number = float(i)
    inverted.append(number)
    
root_of_next_psquare = int(len(inverted) ** 0.5)



```

Reshaping the matrix and getting the original matrix

```python
matrix = np.array(inverted).reshape(root_of_next_psquare,root_of_next_psquare)
print(f'{matrix} \n')
original_matrix = np.linalg.inv(matrix)

original_matrix = original_matrix.astype(int)
print(f'{original_matrix} \n')
message_list = np.array(original_matrix).flatten()
```
