```python
import numpy as np
```

Fetching the data

```python
file = open('./chatlog.txt','r')
data = file.read()
file.close()
```

Converting the data into suitable format

```python
len_of_data = len(data)
root_of_next_psquare = int(len_of_data**0.5) + 1
perfect_square = root_of_next_psquare ** 2


if len(data) < perfect_square:
    adder = perfect_square - len(data)
    for i in range(0,adder):
        data = data + ' '

print(len(data))
```

```python
##Random Number generator
import random
def random_generator(limit):
    li = [2,4,6,8]
    n = 0
    n += random.choice(li)
    for i in range(0,limit-1):
        n *= 10
        n += random.choice(li)
    return n
```

Converting into matrix by multiplying


```python
ascii_list = list(data.encode('ascii'))
print(f'{ascii_list} \n')
def number_separator(n):
    n = str(n)
    li = []
    for i in n:
        li.append(i)
    for item in li:
        item = int(item)
    return li

n = random_generator(root_of_next_psquare)
li = number_separator(n)
for i in range(0,root_of_next_psquare):
    for j in range(0,root_of_next_psquare):
        ascii_list[root_of_next_psquare*i+j] *= int(li[i])
        

matrix = np.array(ascii_list).reshape(root_of_next_psquare,root_of_next_psquare)
print(f'{matrix} \n')
```

Inverting the matrix and preparing the transmission


```python
ivnverse_matrix = np.linalg.inv(matrix)
final_transmission_list = list(i for j in ivnverse_matrix for i in j)
```
